# pandasによるデータの可視化




```python
import pandas as pd
import warnings
warnings.simplefilter('ignore')

#例　アヤメのデータ
df = pd.read_csv('http://logopt.com/data/iris.data', names=['sepal length','sepal width','petal length','petal width', 'class'])
df.head()
```

ヒストグラフ（histメソッド）


*   数値データの分布の概要を知る際に便利



```python
df.hist(bins=10)
```

# 散布図(plot.scatterメソッド)


*   2つの数値データの関係を知る際に便利




```python
df.columns
df.plot.scatter(x="sepal length",y="petal length")
```

# 面積図（plot.areaメソッド）

```python
df.plot.area() 
```

# 棒グラフ (plot.barメソッド)

```python
df.plot.bar( )
```

# 箱ひげ図(plot.boxメソッド)

```python
df.plot.box()
```

# カーネル密度推定(plot.densityメソッド)

```python
df.plot.density( )
```

#6角形格子図(plot.hexbinメソッド)

```python
df.plot.hexbin(x='sepal length', y= 'petal length', gridsize=30)
```

#線グラフ（plot.lineメソッド）

```python
df.plot.line( )
```
